```python
import pandas as pd

version = 'corr_diff'
df = []
corr_dict = load_dict('/Volumes/BPlus/primeRec/results/%s.pickle'%version)
for sub in corr_dict:
    d = [sub]
    d.extend([np.mean(corr) for corr in corr_dict[sub] ])
    df.append(d)
print(df)

df = pd.DataFrame(df,  columns=['SUBJ', 'All self','All other', 
                                'SUP self','SUP other', 
                                'ENH self','ENH other'])

df.to_csv('%s.csv'%version, index=False, encoding='cp949') 
```
