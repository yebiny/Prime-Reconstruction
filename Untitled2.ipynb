```python
data_path = 'data/celeba_split/'
data_list = os.listdir(data_path)
x_data = np.load(data_path+data_list[0])
x_data.shape
```

```python
def plot_recimg(model, image):
    org_img = image[:100]
    rec_img = model(image[:100], training=False)
    fig, ax = plt.subplots(6, 10, figsize=(20, 10))
    for i in range(10):
        for j in range(6):
            if j%2 ==0: img=org_img 
            else: img=rec_img
            ax[j][i].set_axis_off()
            ax[j][i].imshow(img[10*(j//2)+i])
    plt.show()
    plt.close('all')
```

```python
plot_recimg(vae, x_data[:1000])
```

```python
h , w = 5, 10
fig, axes = plt.subplots(h, w, figsize=(w*2, h*2),
                         subplot_kw={'xticks': [], 'yticks': []})
rec_img = vae(x_data[:100])
org_img = x_data[:100]
for ax, i in zip(axes.flat, range(w*h)):
    if i%2 ==0: img = org_img[i//2]
    else: img = rec_img[i//2]
    ax.imshow(img)
    ax.set_title(i//2)
plt.show()
```
